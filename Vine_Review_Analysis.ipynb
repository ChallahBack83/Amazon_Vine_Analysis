```python
import pandas as pd
import numpy as np
```

```python
# Read in vine_table
vine_data = "Resources/vine_table.csv"

vine_review_df = pd.read_csv(vine_data)
vine_review_df.head()
```

```python
count = vine_review_df["review_id"].count()
print(count)
```

```python
#1 Filter into new df where total_votes count is equal or greater than 20
vine_votes_df = vine_review_df[vine_review_df.total_votes >= 20]
vine_votes_df.head(10)
```

```python
count2 = vine_votes_df["review_id"].count()
print(count2)
```

```python
#2 filter new_df: retrieve rows where number of  helpful_votes/total_votes => 50%

# find percentage 
vine_votes_df2 = vine_votes_df.copy()
vine_votes_df2['percent'] = round((vine_votes_df2['helpful_votes']/vine_votes_df2['total_votes'])*100,2)
vine_votes_df2.head(10)

```

```python
# Filter where >= 50%

high_votes = vine_votes_df2[vine_votes_df.percent >= 50]
high_votes.head(10)
```

```python
high_vote_count = high_votes["review_id"].count()
high_vote_count
```

```python
#3 filter new_df2: retrieve all rows where review was part of Vine program(paid) "vine"=='Y'
vine_reviews = high_votes[high_votes.vine == 'Y']
vine_reviews.head(10)
```

```python
#4 filter new_df2: retrieve all rows where review was NOT part of Vine(unpaid) "vine"=='N'
no_vine_reviews = high_votes[high_votes.vine == 'N']
no_vine_reviews.head(10)
```

```python
#5 determine total # PAID reviews, the # of 5-star reviews, and % of 5 star reviews for both PAID & UNPAID

# PAID vs UNPAID
total_reviews = high_votes.groupby('vine').count()
total_reviews
```

```python
# 5 star PAID vs UNPAID
five_star =high_votes.loc[high_votes['star_rating'] == 5.0].groupby('vine').count()
five_star["review_id"]
```

```python
# % of 5 star reviews PAID vs UNPAID

perc_5star = round((five_star['review_id']/total_reviews['review_id'])*100,2)
perc_5star
```

```python
# total percentage of paid vs unpaid
perc_total = round((total_reviews['review_id']/high_vote_count)*100,2)
perc_total
```
